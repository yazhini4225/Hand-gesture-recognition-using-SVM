```python
data = []
labels = []
IMG_SIZE = 64

print(f"Starting data loading from: {path}")

subject_folders_found = False
for subject_folder in os.listdir(path):
    subject_folders_found = True
    subject_path = os.path.join(path, subject_folder)
    if os.path.isdir(subject_path):
        gesture_folders_found = False
        for gesture_folder in os.listdir(subject_path):
            gesture_folders_found = True
            gesture_path = os.path.join(subject_path, gesture_folder)
            if os.path.isdir(gesture_path):
                images_found = False
                for img_name in os.listdir(gesture_path):
                    images_found = True
                    img_path = os.path.join(gesture_path, img_name)
                    img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
                    if img is not None:
                        img = cv2.resize(img, (IMG_SIZE, IMG_SIZE))
                        data.append(img.flatten())
                        labels.append(gesture_folder)
print(f"Loaded samples during processing: {len(data)}")
```

```python
print("Loaded samples:", len(data))
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test= train_test_split(data, labels, test_size=0.2, random_state=42)
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report
```

```python
svm_linear=SVC(kernel="linear")
svm_linear.fit(x_train,y_train)
y_pred=svm_linear.predict(x_test)
```

```python
print(accuracy_score(y_test,y_pred))
```

```python
svm_poly=SVC(kernel="poly")
svm_poly.fit(x_train,y_train)
y_pred=svm_poly.predict(x_test)
print(accuracy_score(y_test,y_pred))
```

```python
svm_rbf=SVC(kernel="rbf")
svm_rbf.fit(x_train,y_train)
y_pred=svm_rbf.predict(x_test)
print(accuracy_score(y_test,y_pred))
```

So, finally the linear svm model gives the best accuracy when compared to others
